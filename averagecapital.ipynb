```python
principal = 800000.0

percent_interest = 5.39
monthly_interest = percent_interest / (100 * 12)

years = 25
payment_months = years * 12
```

```python
for i in range(payment_months):
    monthly_payment = principal / payment_months
    interest = (principal - monthly_payment * i) * monthly_interest
    print('%03d ￥%.2f' % (i + 1, interest + monthly_payment))
```

```python
total_interest = (payment_months + 1) * principal * monthly_interest / 2
print(total_interest)
```

```python
total_cost = principal + total_interest
print(total_cost)
```
